```
#import library
import pandas as pd 
import seaborn as sns
from matplotlib import pyplot as plt
colors = ['#1f77b4', '#aec7e8', '#ff7f0e', '#800000', '#2ca02c', '#98df8a',	'#FF0000']

plt.rcdefaults()
plt.style.use('fivethirtyeight')

#coba
#import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
```

```
#inisialisasi variabel
data = pd.read_csv('selectedseaconfirmated.csv',engine='python')

daycount = data['Day']
MY = data['Malaysia']
SG = data['Singapore']
IN = data['Indonesia']
BR = data['Brunei']
VI = data['Vietnam']
TH = data['Thailand']
PH = data['Philippines']

```

```
#memvisualisasikan line graf menggunakan library matplotlib
plt.plot(daycount,MY,color=colors[0],label='Malaysia') #->mengakses data
plt.plot(daycount,SG,color=colors[1],label='Singapore')#->mengakses data
plt.plot(daycount,IN,color=colors[2],label='Indonesia')#->mengakses data
plt.plot(daycount,BR,color=colors[3],label='Brunei')#->mengakses data
plt.plot(daycount,VI,color=colors[4],label='Vietnam')#->mengakses data
plt.plot(daycount,TH,color=colors[5],label='Thailand')#->mengakses data
plt.plot(daycount,PH,color=colors[6],label='Philippines')#->mengakses data

plt.legend()
plt.title('Corona virus infeksi Asia tenggara days 0-300')
plt.ylabel('Jumlah terinfeksi')#buat label y
plt.xlabel('Jumlah Hari')#buat label x


plt.show()
```

```
#sns data->membat data seaborn(heatmap) = menampilkan data dengan representasi warna yang berbeda-beda.
#menggunakan library seaborn
corelation = data.corr()
sns.heatmap(corelation,xticklabels=corelation.columns,yticklabels=corelation.columns
            ,annot=True)
```

```
#piechart
country=['Brunei','Malaysia', 'Singapore','Vietnam', 'Indonesia',  'Thailand','Philippines'] #->memasukkan data negara
case=[150,58847,58183,1316,506302,3296,421722] #->memasukkan data jumlah infeksi
colors=['black','gold','yellowgreen','black','red','black','green']
explode=[0.1,0,0,0.1,0,0.1,0]


plt.pie(case,labels=country,explode=explode,autopct='%0.2f%%',colors=colors)

plt.title('pieChart ASEAN')
plt.tight_layout()
plt.show()
```

```
#barchart
country =['Malaysia', 'Singapore', 'Indonesia', 'Brunei', 'Vietnam','Thailand','Philippines']
y_pos = np.arange(len(country))
case = [58847,58183,506302,150,1316,3296,421722]

plt.barh(y_pos, case, align='center', alpha=0.9)
plt.yticks(y_pos, country)
plt.xlabel('Infections')
plt.title('Case Infection BarChart')

plt.show()
```
